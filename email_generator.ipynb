```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/software-engineer-ii-salesforce-remote-work-option/job/R-56234")

page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import PromptTemplate

# prompt_template = ChatPromptTemplate([
#     ("system", "You are a helpful assistant"),
#     ("user", "Tell me a joke about {topic}")
# ])

# prompt_template.invoke({"topic": "cats"})


# prompt_extract = PromptTemplate.from_template("Tell me a joke about {topic}")
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPE TEXT FROM WEBSITE:
    {page_data} 
    ### INSTRUCTION
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON object containing 
    following keys: 'role', 'experience', 'skills' and 'description'.
    Do not return a list. Only return a single JSON object.
    ### VALID JSON (NO PREAMBLE):
    """
)



# prompt_template.invoke({"topic": "cats"})
chain_extract = prompt_extract | llm
res = chain_extract.invoke(input= {'page_data': page_data} )

print(res.content)

```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import pandas as pd
df = pd.read_csv("porfolio.csv")
display(df)
```

```python
df.drop(columns="Unnamed: 2")
```

```python
import chromadb
import uuid

client = chromadb.PersistentClient('Vectorstore')
collection = client.get_or_create_collection(name="portfolio_data")

if not collection.count():
    for _,row in df.iterrows():
        collection.add(documents = row['Techstack'],
                       metadatas={"links" : row['Links']},
                       ids = [str(uuid.uuid4())]
)
```

```python
links = collection.query(query_texts=job['skills'], n_results=2).get('metadatas', [])
links
```

```python
results = collection.query(
    query_texts=["Experience in Pthon", "Expertise in React"], # Chroma will embed this for you
    n_results=2 # how many results to return
).get('metadatas',[])
print(results)
```
